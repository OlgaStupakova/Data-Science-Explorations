````python

query_collision_counts = '''
    SELECT
    EXTRACT(MONTH FROM collision_date) AS month,
    COUNT(*) AS collision_count
FROM
    collisions
GROUP BY
    month
ORDER BY
    collision_count DESC;
'''

collision_counts_df = pd.read_sql(query_collision_counts, engine)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(collision_counts_df['month'], collision_counts_df['collision_count'], color='skyblue')
plt.xlabel('Month')
plt.ylabel('Collision Count')
plt.title('Number of Collisions by Month')
plt.xticks(range(1, 13)) 
plt.show()
````

````python
# Convert data to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_train_encoded.values)
y_train_tensor = torch.LongTensor(y_train.values)
X_val_tensor = torch.FloatTensor(X_val_encoded.values)
y_val_tensor = torch.LongTensor(y_val.values)
````

````python
input_size = X_train_encoded.shape[1]
import torch.nn.functional as F

class CustomModel(nn.Module):
    def __init__(self, input_size, hidden_sizes, output_size, activation=F.relu, dropout_rate=0.5):
        super(CustomModel, self).__init__()

        self.input_layer = nn.Linear(input_size, hidden_sizes[0])
        self.hidden_layers = nn.ModuleList()

        for i in range(1, len(hidden_sizes)):
            self.hidden_layers.append(nn.Linear(hidden_sizes[i-1], hidden_sizes[i]))

        self.output_layer = nn.Linear(hidden_sizes[-1], output_size)
        self.dropout = nn.Dropout(dropout_rate)
        self.activation = activation

    def forward(self, x):
        x = self.activation(self.input_layer(x))

        for i in range(len(self.hidden_layers)):
            x = self.dropout(x)
            x = self.activation(self.hidden_layers[i](x))

        x = self.output_layer(x)
        return x
````

````python
# Model 1: One hidden layer with ReLU activation
model1 = CustomModel(input_size, [256], output_size=2, activation=F.relu, dropout_rate=0.5)

# Model 2: Two hidden layers with Tanh activation
model2 = CustomModel(input_size, [256, 128], output_size=2, activation=F.tanh, dropout_rate=0.5)

# Model 3: Three hidden layers with Sigmoid activation
model3 = CustomModel(input_size, [256, 128, 64], output_size=2, activation=F.sigmoid, dropout_rate=0.5)

````

````python
from torch.utils.data import DataLoader

# hyperparameters
learning_rate = 0.001
num_epochs = 100
batch_size = 16
epoch_step = 20

# loss function and the optimizer
criterion = nn.CrossEntropyLoss()

# DataLoader for training and validation
train_loader = DataLoader(TensorDataset(X_train_tensor, y_train_tensor), batch_size=batch_size, shuffle=True)
val_loader = DataLoader(TensorDataset(X_val_tensor, y_val_tensor), batch_size=batch_size)

# Training and evaluation function
def train_and_evaluate(model, train_loader, val_loader):
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    best_f1 = 0.0

    for epoch in range(0, num_epochs, epoch_step):
        model.train()
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

        if epoch % 10 == 0:
            # Print training loss at each epoch
            print(f'Epoch [{epoch}/{num_epochs}], Loss: {loss.item():.4f}')

        # Evaluation on validation set
        model.eval()
        with torch.no_grad():
            val_outputs = model(X_val_tensor)
            val_loss = criterion(val_outputs, y_val_tensor)
            _, val_predictions = torch.max(val_outputs, 1)

        # F1 score on validation set
        f1 = f1_score(y_val_tensor.numpy(), val_predictions.numpy())
        print(f'Epoch [{epoch}/{num_epochs}], Validation Loss: {val_loss.item():.4f}, F1 Score: {f1:.4f}')

        # Save the model with the best F1 score
        if f1 > best_f1:
            best_f1 = f1
            best_model = model.state_dict()

    # final metrics and model parameters
    print(f'Final F1 Score: {best_f1:.4f}')
    return best_model
````
